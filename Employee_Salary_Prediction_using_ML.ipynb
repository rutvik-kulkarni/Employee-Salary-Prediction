```python
from sklearn.preprocessing import LabelEncoder   #import libarary
encoder=LabelEncoder()                       #create object
data['workclass']=encoder.fit_transform(data['workclass']) #7 categories   0,1, 2, 3, 4, 5, 6,
data['marital-status']=encoder.fit_transform(data['marital-status'])   #3 categories 0, 1, 2
data['occupation']=encoder.fit_transform(data['occupation'])
data['relationship']=encoder.fit_transform(data['relationship'])      #5 categories  0, 1, 2, 3, 4
data['race']=encoder.fit_transform(data['race'])
data['gender']=encoder.fit_transform(data['gender'])    #2 catogories     0, 1
data['native-country']=encoder.fit_transform(data['native-country'])
```

```python
data
```

```python
x=data.drop(columns=['income'])
y=data['income']
x
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import (
    RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier,
    ExtraTreesClassifier
)
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
import joblib

try:
    from xgboost import XGBClassifier
    xgboost_available = True
except ImportError:
    xgboost_available = False
    print("XGBoost is not installed. Skipping XGBClassifier.")

# Encode labels to ensure all models return consistent outputs
le = LabelEncoder()
y = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "NaiveBayes": GaussianNB(),
    "AdaBoost": AdaBoostClassifier(),
    "ExtraTrees": ExtraTreesClassifier()
}

if xgboost_available:
    models["XGBoost"] = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

results = {}

# Train and evaluate using pipelines with scaling
for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))

# Get best model
best_model_name = max(results, key=results.get)
best_model_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', models[best_model_name])
])
best_model_pipeline.fit(X_train, y_train)

print(f"\nBest model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save best model
joblib.dump(best_model_pipeline, "best_model_pipeline.pkl")
print("Saved best model pipeline as best_model_pipeline.pkl")

# Save label encoder if XGBoost was used
if best_model_name == "XGBoost":
    joblib.dump(le, "label_encoder.pkl")
    print("Saved label encoder as label_encoder.pkl")
```

```python
import gradio as gr
import pandas as pd
import joblib

# Load trained model
model = joblib.load("best_model_pipeline.pkl")

# Try loading label encoder (only needed if model returns 0/1)
try:
    le = joblib.load("label_encoder.pkl")
    use_label_encoder = True
except:
    le = None
    use_label_encoder = False

# Encoding mappings from training
workclass_map = {'Private': 4, 'Self-emp-not-inc': 5, 'Local-gov': 2, 'State-gov': 6,
                 'Federal-gov': 1, 'Self-emp-inc': 3, 'Others': 0}
occupation_map = {'Prof-specialty': 8, 'Exec-managerial': 3, 'Adm-clerical': 0, 'Craft-repair': 2,
                  'Sales': 9, 'Other-service': 7, 'Machine-op-inspct': 5, 'Transport-moving': 11,
                  'Handlers-cleaners': 4, 'Tech-support': 10, 'Farming-fishing': 1,
                  'Protective-serv': 6, 'Priv-house-serv': 12, 'Armed-Forces': 13, 'Others': 14}
gender_map = {'Male': 1, 'Female': 0}
marital_map = {'Married-civ-spouse': 1, 'Never-married': 2, 'Divorced': 0}
relationship_map = {'Husband': 0, 'Not-in-family': 3, 'Own-child': 4,
                    'Unmarried': 1, 'Wife': 2}
race_map = {'White': 4, 'Black': 0, 'Asian-Pac-Islander': 1,
            'Amer-Indian-Eskimo': 2, 'Other': 3}
country_map = {'United-States': 38, 'Mexico': 24, 'Philippines': 30, 'Germany': 10,
               'Canada': 4, 'India': 14, 'England': 11, 'Cuba': 6, 'Others': 0}

column_order = ['age', 'workclass', 'marital-status', 'occupation', 'relationship',
                'race', 'gender', 'capital-gain', 'hours-per-week', 'native-country', 'educational-num']


def predict_salary(age, educational_num, workclass, occupation, hours_per_week,
                   capital_gain, gender, marital_status, relationship, race, native_country):
    try:
        input_data = pd.DataFrame([{
            'age': age,
            'workclass': workclass_map.get(workclass, 0),
            'fnlwgt': 200000,
            'educational-num': educational_num,
            'marital-status': marital_map.get(marital_status, 0),
            'occupation': occupation_map.get(occupation, 14),
            'relationship': relationship_map.get(relationship, 0),
            'race': race_map.get(race, 4),
            'gender': gender_map.get(gender, 1),
            'capital-gain': capital_gain,
            'capital-loss': 0,
            'hours-per-week': hours_per_week,
            'native-country': country_map.get(native_country, 0)
        }])

        expected_cols = ['age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status',
                         'occupation', 'relationship', 'race', 'gender',
                         'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']
        input_data = input_data[expected_cols]

        prediction = model.predict(input_data)[0]
        if use_label_encoder:
            prediction = le.inverse_transform([prediction])[0]

        return f"Prediction: {prediction}"

    except Exception as e:
        print("ERROR:", str(e))
        return f"Error: {str(e)}"


def batch_predict(file):
    df = pd.read_csv(file)

    df['workclass'] = df['workclass'].map(workclass_map).fillna(0).astype(int)
    df['occupation'] = df['occupation'].map(occupation_map).fillna(14).astype(int)
    df['gender'] = df['gender'].map(gender_map).fillna(1).astype(int)
    df['marital-status'] = df['marital-status'].map(marital_map).fillna(0).astype(int)
    df['relationship'] = df['relationship'].map(relationship_map).fillna(0).astype(int)
    df['race'] = df['race'].map(race_map).fillna(4).astype(int)
    df['native-country'] = df['native-country'].map(country_map).fillna(0).astype(int)

    df = df[column_order]
    preds = model.predict(df)
    if use_label_encoder:
        preds = le.inverse_transform(preds)
    df['Prediction'] = preds
    return df


with gr.Blocks(title="Employee Salary Prediction") as demo:
    gr.Markdown("# Employee Salary Prediction")
    gr.Markdown("Fill out the details below to predict whether an employee earns >50K or 50K.")

    with gr.Row():
        with gr.Column():
            age = gr.Slider(17, 75, label="Age")
            educational_num = gr.Slider(5, 16, label="Education Level (Numerical)")
            workclass = gr.Dropdown(choices=list(workclass_map.keys()), label="Workclass")
            occupation = gr.Dropdown(choices=list(occupation_map.keys()), label="Occupation")
            hours_per_week = gr.Slider(1, 80, label="Hours per Week")
            capital_gain = gr.Number(label="Capital Gain", value=0)

        with gr.Column():
            gender = gr.Radio(choices=list(gender_map.keys()), label="Gender")
            marital_status = gr.Dropdown(choices=list(marital_map.keys()), label="Marital Status")
            relationship = gr.Dropdown(choices=list(relationship_map.keys()), label="Relationship")
            race = gr.Dropdown(choices=list(race_map.keys()), label="Race")
            native_country = gr.Dropdown(choices=list(country_map.keys()), label="Native Country")

    submit_btn = gr.Button("Predict Salary Class")
    result = gr.Textbox(label="Prediction")

    submit_btn.click(fn=predict_salary,
                     inputs=[age, educational_num, workclass, occupation, hours_per_week,
                             capital_gain, gender, marital_status, relationship, race, native_country],
                     outputs=result)

    gr.Markdown("---")
    gr.Markdown("## Batch Prediction")
    csv_file = gr.File(label="Upload CSV File", file_types=[".csv"])
    output_df = gr.Dataframe(label="Predictions")

    csv_file.change(fn=batch_predict, inputs=csv_file, outputs=output_df)

demo.launch()
```
